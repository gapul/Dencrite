```python
# 必要なライブラリをインポート
import pandas as pd

# CSVファイルを読み込む
df = pd.read_csv('greek_tweets_labeled.csv')

# 削除前のデータ数を表示
print("削除前のデータ数:", len(df))

# TweetまたはSourceが空のデータを削除
# dropna()を使用して、Tweet列またはSource列が空のデータを削除
# subset: 確認する列を指定
# how='any': 指定した列のいずれかが空の場合に行を削除
df_cleaned = df.dropna(subset=['Tweet', 'Source'], how='any')

# 空文字列も削除
df_cleaned = df_cleaned[
    (df_cleaned['Tweet'].str.strip() != '') &
    (df_cleaned['Source'].str.strip() != '')
]

# 削除後のデータ数を表示
print("削除後のデータ数:", len(df_cleaned))

# 削除されたデータ数を表示
print("削除されたデータ数:", len(df) - len(df_cleaned))

# 結果の最初の数行を表示
df_cleaned.head()
```

```python
# 必要なライブラリをインポート
import pandas as pd

# 前回のクリーニング済みデータを使用
# Sourceの出現回数を確認
source_counts = df_cleaned['Source'].value_counts()

# 上位3種類のSourceを表示
print("上位3種類のSource:")
print(source_counts.head(3))

# 上位3種類のSourceを取得
top_3_sources = source_counts.head(3).index.tolist()

# 上位3種類のSourceのデータのみを抽出
df_top_sources = df_cleaned[df_cleaned['Source'].isin(top_3_sources)]

# 処理前後のデータ数を表示
print("\nSource絞り込み前のデータ数:", len(df_cleaned))
print("Source絞り込み後のデータ数:", len(df_top_sources))
print("削除されたデータ数:", len(df_cleaned) - len(df_top_sources))

# Sourceごとのデータ数を表示
print("\n絞り込み後のSourceごとのデータ数:")
print(df_top_sources['Source'].value_counts())

# 結果の最初の数行を表示
print("\nデータサンプル:")
df_top_sources.head()

df_cleaned = df_top_sources
```

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# データの確認
print("Unique values in Source column:")
print(df_cleaned['Source'].unique())

# Calculate Tweet length and create new column
df_cleaned['Tweet_length'] = df_cleaned['Tweet'].str.len()

# Create platform mapping
platform_mapping = {
    'Android': 'Mobile',
    'iOS': 'Mobile',
    'Web App': 'PC'
}

# Create new column with mapped platforms
df_cleaned['Platform'] = df_cleaned['Source'].map(platform_mapping)

# マッピング後の確認
print("\nUnique values in Platform column:")
print(df_cleaned['Platform'].unique())

# null値の確認
print("\nNull values in Platform column:")
print(df_cleaned['Platform'].isnull().sum())

# 各プラットフォームのデータ数を確認
print("\nCount of records by Platform:")
print(df_cleaned['Platform'].value_counts())
```

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate Tweet length and create new column
df_cleaned['Tweet_length'] = df_cleaned['Tweet'].str.len()

# Create platform mapping with correct source names
platform_mapping = {
    'Twitter for Android': 'Mobile',
    'Twitter for iPhone': 'Mobile',
    'Twitter Web App': 'PC'
}

# Create new column with mapped platforms
df_cleaned['Platform'] = df_cleaned['Source'].map(platform_mapping)

# Verify the mapping
print("Unique values in Platform column after correction:")
print(df_cleaned['Platform'].unique())
print("\nCount of records by Platform:")
print(df_cleaned['Platform'].value_counts())

# Calculate basic statistics for each Platform
length_stats = df_cleaned.groupby('Platform')['Tweet_length'].agg([
    ('Mean', 'mean'),
    ('Variance', 'var'),
    ('Standard_Deviation', 'std'),
    ('Minimum', 'min'),
    ('Maximum', 'max')
]).round(2)

print("\nTweet length statistics by Platform:")
print(length_stats)

# Visualize distributions
plt.figure(figsize=(15, 6))

# 1. Box Plot
plt.subplot(1, 2, 1)
sns.boxplot(x='Platform', y='Tweet_length', data=df_cleaned)
plt.title('Tweet Length Distribution (Box Plot)')
plt.xticks(rotation=45)
plt.ylabel('Tweet Length')

# 2. Violin Plot
plt.subplot(1, 2, 2)
sns.violinplot(x='Platform', y='Tweet_length', data=df_cleaned)
plt.title('Tweet Length Distribution (Violin Plot)')
plt.xticks(rotation=45)
plt.ylabel('Tweet Length')

plt.tight_layout()
plt.show()

# Draw histograms (separate graphs for each Platform)
plt.figure(figsize=(10, 5))
for i, platform in enumerate(df_cleaned['Platform'].unique(), 1):
    plt.subplot(1, 2, i)
    platform_data = df_cleaned[df_cleaned['Platform'] == platform]['Tweet_length']
    sns.histplot(data=platform_data, bins=30, kde=True)
    plt.title(f'Tweet Length Distribution for {platform}')
    plt.xlabel('Tweet Length')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Display detailed descriptive statistics
for platform in df_cleaned['Platform'].unique():
    print(f"\nDetailed statistics for {platform}:")
    platform_data = df_cleaned[df_cleaned['Platform'] == platform]['Tweet_length']

    # Calculate quartiles
    q1 = np.percentile(platform_data, 25)
    q3 = np.percentile(platform_data, 75)
    iqr = q3 - q1

    stats = {
        'Count': len(platform_data),
        'Mean': np.mean(platform_data),
        'Median': np.median(platform_data),
        'Variance': np.var(platform_data),
        'Standard_Deviation': np.std(platform_data),
        'First_Quartile': q1,
        'Third_Quartile': q3,
        'IQR': iqr,
        'Skewness': platform_data.skew(),
        'Kurtosis': platform_data.kurtosis()
    }

    for metric, value in stats.items():
        print(f"{metric}: {value:.2f}")
```

```python
# 必要なライブラリをインポート
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Tweetの長さを計算して新しい列を作成
df_cleaned['Tweet_length'] = df_cleaned['Tweet'].str.len()

# Sourceごとの基本統計量を計算
length_stats = df_cleaned.groupby('Source')['Tweet_length'].agg([
    ('平均値', 'mean'),
    ('分散', 'var'),
    ('標準偏差', 'std'),
    ('最小値', 'min'),
    ('最大値', 'max')
]).round(2)

print("Sourceごとのツイート長の統計情報:")
print(length_stats)

# 分布の可視化
plt.figure(figsize=(15, 6))

# 1. Box Plot
plt.subplot(1, 2, 1)
sns.boxplot(x='Source', y='Tweet_length', data=df_cleaned)
plt.title('ツイート長の分布 (Box Plot)')
plt.xticks(rotation=45)
plt.ylabel('ツイート長')

# 2. Violin Plot
plt.subplot(1, 2, 2)
sns.violinplot(x='Source', y='Tweet_length', data=df_cleaned)
plt.title('ツイート長の分布 (Violin Plot)')
plt.xticks(rotation=45)
plt.ylabel('ツイート長')

plt.tight_layout()
plt.show()

# ヒストグラムの描画（Sourceごとに別々のグラフ）
plt.figure(figsize=(15, 5))
for i, source in enumerate(df_cleaned['Source'].unique(), 1):
    plt.subplot(1, 3, i)
    source_data = df_cleaned[df_cleaned['Source'] == source]['Tweet_length']
    sns.histplot(data=source_data, bins=30, kde=True)
    plt.title(f'{source}のツイート長分布')
    plt.xlabel('ツイート長')
    plt.ylabel('頻度')

plt.tight_layout()
plt.show()

# 記述統計の詳細表示
for source in df_cleaned['Source'].unique():
    print(f"\n{source}の詳細統計:")
    source_data = df_cleaned[df_cleaned['Source'] == source]['Tweet_length']

    # 四分位数を計算
    q1 = np.percentile(source_data, 25)
    q3 = np.percentile(source_data, 75)
    iqr = q3 - q1

    stats = {
        '件数': len(source_data),
        '平均値': np.mean(source_data),
        '中央値': np.median(source_data),
        '分散': np.var(source_data),
        '標準偏差': np.std(source_data),
        '第1四分位数': q1,
        '第3四分位数': q3,
        'IQR': iqr,
        '歪度': source_data.skew(),
        '尖度': source_data.kurtosis()
    }

    for metric, value in stats.items():
        print(f"{metric}: {value:.2f}")
```
